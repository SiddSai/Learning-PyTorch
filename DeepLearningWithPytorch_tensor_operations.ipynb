```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape Data: Reshape and View

my_torch = my_torch.reshape(5, 2)
my_torch
```

```python
# Reshape if we do not know # items. Use -1
my_torch2 = torch.arange(15)
```

```python
my_torch2 = my_torch2.reshape(-1, 5)
my_torch2
```

```python
# Try view

my_torch3 = torch.arange(15)
my_torch3
```

```python
my_torch4 = my_torch3.view(3, 5)
my_torch4
```

```python
# with reshape and view, they will update
my_torch5 = torch.arange(10)
my_torch5
```

```python
my_torch6 = my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1] = 4141
print(my_torch5)
print(my_torch6)
```

```python
# Slices
my_torch7 = torch.arange(10)
my_torch7
```

```python
# Grab a specific item
my_torch7[7]
```

```python
# Grab a slice
my_torch8 = my_torch7.reshape(5, 2)
my_torch8
```

```python
my_torch8[:,1]
```

```python
# Return column
my_torch8[:,1:]
```
