Neural Network:
- Way to learn with machines
- Similar to brain with neurons (how they developed)
- Starts in input layer
- Ends in input layer
- Learning happens in the layers of neurons
- Layers have different weights (probablistic weights)
- Different types of NN's have different learning methods through neurons
- Workflow: Input -> Learning -> Output


```python
import torch
import numpy as np
```

Tensor:


* Like a numpy array or python list
* Machine Learning: numpy list
* Deep Learning: tensors (more powerful/scaled for deep learning)
* Multi-dimensional matrix containing elements of a single data type
* Default data type: float32




```python
### Lists

# Test
my_list = [[1, 2, 3, 4], [5, 6, 7, 8]]
print(my_list)
```

```python
### Numpy Arrays

# Test
np1 = np.random.rand(3, 4)
print(np1)

# Check data type
print(np1.dtype)

```

```python
### Tensors

# 3D
tensor_2d = torch.randn(3, 4)
print(tensor_2d)
tensor_2d.dtype
```

```python
# 3D
tensors_3d = torch.zeros(2, 3, 4)
print(tensors_3d)
```

```python
# Convert numpy array to tensor

np_to_tensor = torch.tensor(np1)
print(np_to_tensor)
```
